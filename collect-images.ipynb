```python
def parse_fname_09(fname):
    s = str(fname).split('/')
    d = {}
    d['name'] = s[0]
    d['date'] = s[1]
    d['time'] = (s[4])[0:8]
    return d

def parse_fname_04_05(fname):
    s = str(fname).split('/')
    d = {}
    d['name'] = s[0]
    d['date'] = s[1]
    d['time'] = f"{s[4]}.{s[5]}.{(s[6])[0:2]}"
    return d
```

```python
p = Path('./cam-09')
filegen = p.glob('**/*.jpg')

#
# for cam-04 & cam-05, format is YYYY-mm-dd/001/jpg/hh/mm/ss{m}[0@0][0].jpg
# for cam-09, format is YYYY-mm-dd/HHhour/jpg/HH.mm.ss[M][0@0][0].jpg
for fname in filegen:
    fname_d = parse_fname_09(fname)
    dest_fname = f"renamed/{fname_d['name']}-{fname_d['date']}-{fname_d['time']}.jpg"
    print(dest_fname)
    !cp $fname $dest_fname
```

```python
#specify ""matplotlib notebook"" for resizeable inline plots:
# works in jupyter notebook
%matplotlib notebook 
# works in vs code; may need to ""pip install ipympl""
#%matplotlib widget
import os
from pathlib import Path
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
import numpy as np
import ipyvolume as ipv
import sklearn.decomposition
```

```python
p = Path('./cam-09')
filegen = p.glob('**/*.jpg')
path = next(filegen)
x  = Image.open(path) 
```

```python
x = np.asarray(ImageOps.grayscale(x))
x_centered = x - np.mean(x)
print(np.mean(x_centered))
plt.figure()
#plt.imshow(x_centered,cmap='gray', vmin=np.min(x_centered), vmax=np.max(x_centered))
plt.imshow(x_centered,cmap='gray')
plt.tight_layout()
plt.show()
```

```python
#
# pca analysis just computes same analysis as SVD 
#
# class sklearn.decomposition.PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', n_oversamples=10, power_iteration_normalizer='auto', random_state=None)[source]
pca = sklearn.decomposition.PCA()
pca.fit(x_centered)
sv = pca.singular_values_
#print(sv)

fig,ax=plt.subplots()
ax.plot(pca.singular_values_,'o')
ax.grid(True)
plt.show()
```

```python
#
# PCA's dimensional reduction is same as using first N eigenvectors
#
pca = sklearn.decomposition.PCA(n_components=25)
pca.fit(x_centered)
y = pca.transform(x_centered)
x2 = pca.inverse_transform(y)

plt.figure()
#plt.imshow(x_centered,cmap='gray', vmin=np.min(x_centered), vmax=np.max(x_centered))
plt.imshow(x2,cmap='gray')
plt.tight_layout()
plt.show()
```

```python
# try pca on a synthetic, toy 3D image
# keep in mind that point of PCA is to reduce dimensionality
# VAR(R[2,:]) is smaller than others
# (so this is a thin disk)
# 
#R = np.eye(2)
NPOINTS=5000
R = np.zeros((3, NPOINTS))
R[0,:] = np.random.normal(loc=0.0, scale=0.75, size=(NPOINTS,))
R[1,:] = np.random.normal(loc=0.0, scale=0.75, size=(NPOINTS,))
R[2,:] = np.random.normal(loc=0.0, scale=0.05, size=(NPOINTS,))


#R = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
print(R.shape)

ipv.quickscatter(R[0,:],R[1,:],R[2,:], size=1, marker="sphere")
```

```python

r_rank = np.linalg.matrix_rank(R)
print(r_rank)
```

```python
#
# fit PCA model to R
# IMPORTANT: shape MUST be samples x dimensions
#
pca = sklearn.decomposition.PCA(n_components = r_rank-1)
pca.fit(R.transpose())

print(pca.n_components_)
print(pca.n_features_)
print(pca.n_samples_)
print(pca.components_.shape)
print(pca.explained_variance_)
print(np.dot(pca.components_[0,:], pca.components_[1,:]))
```

```python
#
# pca.transform transforms 3D data into 2D data 
#
Rt = pca.transform(R.transpose())
print(Rt.shape)
```

```python
#
# plot 2D shape (3rd dimension is eliminated via projection to plane)
fig,ax=plt.subplots()
ax.plot(Rt[:,0],Rt[:,1],'.')
plt.show()
```
